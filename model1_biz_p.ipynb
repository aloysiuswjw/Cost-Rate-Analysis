<h1>Model 1: Business Performance from 2002 to 2022 (Historial Data Analysis) </h1>

```python
from IPython.core.interactiveshell import InteractiveShell
from seaborn import heatmap
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
import numpy as np
import datetime
from datetime import datetime # for strptime
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import seaborn as sns
import math # for rounding up values
%matplotlib inline
sns.set()
pd.options.display.max_rows = 4000

df = pd.read_csv('CR_COST_1.csv', converters={'PARTY_ID' : str, 'PORT_ID': str,  'CNTR_TYPE': str})
df['ETD_POL_D'] = pd.to_datetime(df['ETD_POL_D'])
df.head(10)
df.info()
```

<h3>Exports freq throughout of the years</h3>

```python
# Group the data by date
grouped_data = df.groupby(df['ETD_POL_D'])

# Unpack to plot
dates, freq = zip(*grouped_data.size().items())

# Convert the dates to a Matplotlib datetime format
dates = [mdates.datestr2num(d.strftime('%Y-%m-%d')) for d in dates]

# Create a figure with a specific size
fig = plt.figure(figsize=(80, 20))

# Set the x-axis date format
ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# Set the tick intervals
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_minor_locator(mdates.MonthLocator())


# Plot the event counts over time
plt.plot(dates[::30], freq[::30], linewidth=2)

plt.xlabel('Date')
plt.ylabel('Number of Events')
plt.title('Event Counts Over Time')
plt.show();
```

```python
# Create a figure with a specific size
fig = plt.figure(figsize=(60, 20))

# Set the x-axis date format
ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# Set the tick intervals
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_minor_locator(mdates.MonthLocator())

# Plot scatterplot
plt.scatter(dates, freq, color='red')

plt.xlabel('Date')
plt.ylabel('Number of Events')
plt.title('Event Counts Over Time')
plt.show();
```
