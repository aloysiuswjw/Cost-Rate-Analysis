```python
import math  # for rounding up values
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime  # for strptime
import datetime
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
from seaborn import heatmap
InteractiveShell.ast_node_interactivity = "all"
%matplotlib inline
sns.set()
pd.options.display.max_rows = 4000

df = pd.read_csv('CR_COST_test.csv', converters={ 'COM_ID': str,
                 'PARTY_ID': str, 'PORT_ID': str, 'ETD_POL_D': str, 'CNTR_TYPE': str})

# Selecting columns needed
df.head()

# Showing the data type for each columns
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)/df.shape[0]
df.info()
```

```python
df1 = df.dropna()
df1.info()
```

```python
# Remove commas from dataframe
df1.COST_RATE = df1.COST_RATE.apply(lambda x: x.replace(',', ''))

# Converting cost rate column to float
df1['COST_RATE'] = df1['COST_RATE'].map(lambda x: float(x))
df1.head()
df1.info()

# Remove negative values
df1 = df1.drop(df1.index[df1['COST_RATE'] < 0])
df1 = df1.drop(df1.index[df1['COST_QTY'] < 0])

# Visualising Data
print("\n")
print("Total of " + str(len(df1)) +
      " rows left after removing rows contain negative cost values.")
df1.head()
```

```python
# Convert float to integer
df1['CNTR_SIZE'] = df1['CNTR_SIZE'].astype(int)
```

```python
# Changing column ETD_POL_D from string to date format day/month/year
df1['ETD_POL_D'] = pd.to_datetime(df1['ETD_POL_D'], errors='coerce')
df1['ETA_POD_D'] = pd.to_datetime(df1['ETA_POD_D'], errors='coerce')

for col in ['COST_C','COST_DESC','COST_UNIT','COST_TERM','POL_ID','VESSEL','PARTY_ID', 'POD_ID','CNTR_TYPE','CNTR_SIZE']:
    df1[col] = df1[col].astype(str)
df1.head()
df1.info()

# Exporting it as new dataframe
df1.to_csv("CR_COST_1.csv", index=False)
```
